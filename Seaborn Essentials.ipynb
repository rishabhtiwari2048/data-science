```python
##FacetGrid
sns.get_dataset_names()
```

```python
iris = sns.load_dataset('iris')
tips = sns.load_dataset('tips')
```

```python
tips.info()
```

```python
iris.info()
```

```python
tips.head()
```

```python
iris.head()
```

```python
g = sns.FacetGrid(data=tips, row='sex',col='smoker',hue='day',size=5, aspect=1.0)
g = g.map(sns.scatterplot,'total_bill','tip')
g.add_legend()
```

```python
g = sns.FacetGrid(data=tips, row='sex',col='time',hue='smoker')
g.map(sns.countplot,'day')
```

```python
##PairGrid
iris.head()
```

```python
g = sns.PairGrid(iris, hue='species', palette='rainbow')
g.map(sns.scatterplot)
g.add_legend()
```

```python
g = sns.PairGrid(tips, hue='day')
g.map_diag(sns.distplot)
g.map_offdiag(sns.scatterplot)
```

```python
g = sns.PairGrid(data=tips, hue='day')
g.map_diag(sns.kdeplot)
g.map_upper(sns.scatterplot)
g.map_lower(sns.lineplot)
```

```python
##JointGrid
g = sns.JointGrid('total_bill','tip',data=tips)
g.plot(sns.regplot,sns.kdeplot)
```
