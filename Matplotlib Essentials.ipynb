# Matplotlib Basics

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x = np.arange(0,11)
y = x ** 2
```

```python
x,y
```

```python
plt.plot(x,y)
plt.xlabel('Numbers')
plt.ylabel('Squares')
plt.title('Numbers and its Square')
```

```python
plt.subplot(1,3,1)
plt.plot(x,y,'rp-')
plt.xlabel('Numbers')
plt.ylabel('Squares')
plt.title('Numbers vs Squares')

plt.subplot(1,3,2)
plt.plot(y,x,'go:')
plt.xlabel('Squares')
plt.ylabel('Numbers')
plt.title('Squares vs Numbers')

plt.subplot(1,3,3)
plt.plot(y,x,'g<--')
plt.plot(x,y,'co-')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Chart')
plt.tight_layout()
```

```python
fig1 = plt.figure()
axis = fig1.add_axes([0.0,0.0,1.0,1.0],projection=None)
axis.plot(x,y)
```

```python
fig2 = plt.figure() #empty canvas
axis = fig2.add_axes([0.0,0.0,1,1]) #set diamensions
axis.plot(x,y,'rp:')
axis.set_xlabel('X Axis')
axis.set_ylabel('Y Axis')
axis.set_title('Title')
plt.show()
```

```python
fig3 = plt.figure()
main_axis = fig3.add_axes([0.0,0.1,1.0,0.9])
inset_axis = fig3.add_axes([0.05,0.4,0.45,0.5])
main_axis.plot(x,y,'r')
inset_axis.plot(x,y,'g')
main_axis.set_xlabel('X Main')
main_axis.set_ylabel('Y Main')
main_axis.set_title('Main Title')
inset_axis.set_xlabel('X Inset')
inset_axis.set_ylabel('Y Inset')
inset_axis.set_title('Subtitle')
```

```python
fig4,axes = plt.subplots()
```

```python
fig5,axes = plt.subplots(nrows=2,ncols=2)
plt.tight_layout()
```

```python
fig6,axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y,'r')
axes[0].set_xlabel('X1')
axes[0].set_ylabel('Y1')
axes[0].set_title('Fig 1')
axes[1].plot(x,y,'g')
axes[1].set_xlabel('X2')
axes[1].set_ylabel('Y2')
axes[1].set_title('Fig 2')
plt.tight_layout()
```

```python
fig, axes = plt.subplots(figsize=(10,10),dpi=200)
axes.plot(x,y,'r--')
axes.set_xlabel('X Axis')
axes.set_ylabel('Y Axis')
axes.set_title('Default')
```

```python
fig.savefig('sample.jpeg',dpi=100)
```

```python
fig1 = plt.figure(figsize=(10,10), dpi = 100)
axes1 = fig1.add_axes([0,0,1,0.4])
axes1.plot(x,x**2,'ro-',label='Square')
axes1.plot(x,x+10,'gp--',label='Line Shift')
axes1.legend(loc=0)
```

```python
fig, axes = plt.subplots(figsize=(7,7),dpi=200)
axes.plot(x,x+1,'r',lw=1,ls='-.',marker='o',mec='black',mew=2,mfc='white',ms=10)
axes.plot(x,x+10,'g',lw=1,ls=':',marker='p',mec='black',mew=2,mfc='white',ms=10)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(10,10), dpi=200)

axes[0].plot(x,x**2,'r',marker='s')
axes[0].set_xlabel('X1')
axes[0].set_ylabel('Y1')
axes[0].set_title('Fig 1')

axes[1].plot(x,x**2,'g',marker='o')
axes[1].set_xlabel('X2')
axes[1].set_ylabel('Y2')
axes[1].set_title('Fig 2')
axes[1].axis('tight')

axes[2].plot(x,x**2,'b',marker='<')
axes[2].set_xlabel('X3')
axes[2].set_ylabel('Y3')
axes[2].set_title('Fig 3')
axes[2].set_xlim((0,10))
axes[2].set_ylim((0,100))
```
